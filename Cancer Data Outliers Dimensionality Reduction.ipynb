### Processing outliers
- Remove (Std deviation or IQR)
- Rescale (log or sqrt)
- Create dummy variable

#### Removing outliers greater than 3 sigma

```python
# high value
am_plus3sd = df.area_mean.mean() +3 * df.area_mean.std()
am_plus3sd
```

```python
# Removing outliers
df1 = df.query('area_mean <= 1710')
df1.info()
```

#### Removing outliers using IQR

```python
# IQR
t = df.area_mean.quantile(0.75) + 1.5 * (df.area_mean.quantile(0.75) - df.area_mean.quantile(0.25))
t
```

```python
df2 = df.query('area_mean <= 1326')
df2.info()
```

### Predictive Analytics: Classification Model


```python
# Changing the target to numeric
df.diagnosis = np.where(df.diagnosis == 'M', 1, 0)
```

```python
# choose predictors
X = df.iloc[:,1:]
```

```python
# target variable
y = df['diagnosis']
```

```python
# import KNN
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 11)
```

```python
# standardizing predictors
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X)
X_std = sc.transform(X)
```

```python
# crossvalidation
from sklearn.model_selection import cross_val_score, StratifiedShuffleSplit
shuffle_split = StratifiedShuffleSplit(test_size = 0.2, n_splits = 10, random_state = 123)
scores = cross_val_score(model, X_std, y, cv = shuffle_split, scoring = 'f1')
print('CV',scores)
print('Mean CV', scores.mean())
```

### Dimensionality Reduction: Principal Component Analysis (PCA)

```python
# correlation matrix
df.corr()
```

```python
# PCA
from sklearn.decomposition import PCA
pca = PCA(n_components = 5)
Xpca = pca.fit_transform(X_std)
```

```python
pca.explained_variance_ratio_
```
